8. Создайте случайную матрицу 10х2 которая представляет собой декартовые координаты и преобразуйте их в полярные

````python
# создаем матрицу 10х2 с рандомными значениями
rand_values = np.random.rand(10, 2)

# делаем разбиение  на координаты
x_dec = rand_values[:, 0]
y_dec = rand_values[:, 1]

# преобразовываем в полярные
r = np.sqrt(pow(x_dec, 2) + pow(y_dec, 2))
phi = np.arctan(y_dec, x_dec)

polar_coords = np.column_stack((r, phi))
print(polar_coords)
````
